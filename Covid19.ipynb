```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'Country/Region': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'Confirmed': [36263, 4880, 27973, 907, 950],
    'Deaths': [1269, 144, 1163, 52, 41],
    'Recovered': [25198, 2745, 18837, 803, 242],
    'Active': [9796, 1991, 7973, 52, 667],
    'New cases': [106, 117, 616, 10, 18],
    'New deaths': [10, 6, 8, 0, 1],
    'New recovered': [18, 63, 749, 0, 0],
    'Deaths / 100 Cases': [3.50, 2.95, 4.16, 5.73, 4.32],
    'Recovered / 100 Cases': [69.49, 56.25, 67.34, 88.53, 25.47],
    'Deaths / 100 Recovered': [5.04, 5.25, 6.17, 6.48, 16.94],
    'Confirmed last week': [35526, 4171, 23691, 884, 749],
    '1 week change': [737, 709, 4282, 23, 201],
    '1 week % increase': [2.07, 17.00, 18.07, 2.60, 26.84],
    'WHO Region': ['Eastern Mediterranean', 'Europe', 'Africa', 'Europe', 'Africa']
}

df = pd.DataFrame(data)

# Convert index to datetime format (replace df.index with df['Date'] if you have a 'Date' column)
df.index = pd.to_datetime('2023-01-01') + pd.to_timedelta(df.index, unit='D')

# Plotting time series data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Confirmed'], label='Confirmed Cases')
plt.plot(df.index, df['Deaths'], label='Deaths')
plt.plot(df.index, df['Recovered'], label='Recovered')
plt.title('COVID-19 Global Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()

```

```python
region_to_plot = 'Europe'
region_df = df[df['WHO Region'] == region_to_plot]

plt.figure(figsize=(12, 6))
plt.plot(region_df.index, region_df['Confirmed'], label='Confirmed Cases in ' + region_to_plot)
plt.plot(region_df.index, region_df['Deaths'], label='Deaths in ' + region_to_plot)
plt.plot(region_df.index, region_df['Recovered'], label='Recovered in ' + region_to_plot)
plt.title(f'COVID-19 Trends in {region_to_plot} Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['New cases'], label='Daily New Cases')
plt.plot(df.index, df['New deaths'], label='Daily New Deaths')
plt.plot(df.index, df['New recovered'], label='Daily New Recovered')
plt.title('Daily COVID-19 Trends')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
countries = df['Country/Region']
confirmed = df['Confirmed']
deaths = df['Deaths']
recovered = df['Recovered']

# Plotting bar chart
plt.figure(figsize=(14, 8))
bar_width = 0.2
index = range(len(countries))
plt.bar(index, confirmed, width=bar_width, label='Confirmed', color='blue')
plt.bar([i + bar_width for i in index], deaths, width=bar_width, label='Deaths', color='red')
plt.bar([i + 2 * bar_width for i in index], recovered, width=bar_width, label='Recovered', color='green')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.title('COVID-19 Cases by Country')
plt.xticks([i + bar_width for i in index], countries, rotation=45, ha='right')
plt.legend()
plt.show()
```

```python
# Sample data (replace this with your actual data)
countries = df['Country/Region']
confirmed = df['Confirmed']
deaths = df['Deaths']
recovered = df['Recovered']

# Calculate percentages
death_percentage = deaths / confirmed * 100
recovered_percentage = recovered / confirmed * 100

# Plotting stacked bar chart
plt.figure(figsize=(14, 8))
plt.bar(countries, confirmed, label='Confirmed', color='blue')
plt.bar(countries, death_percentage, bottom=confirmed, label='Deaths (%)', color='red')
plt.bar(countries, recovered_percentage, bottom=confirmed + death_percentage, label='Recovered (%)', color='green')
plt.xlabel('Countries')
plt.ylabel('Percentage')
plt.title('Percentage of Deaths and Recovered Cases Compared to Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
